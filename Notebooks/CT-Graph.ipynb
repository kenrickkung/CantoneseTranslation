```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
csv = pd.read_csv("Loss/opus-baseline.csv")
csv
```

```python
def plot_loss(input_path, output_path, name):
    csv = pd.read_csv(input_path)

    epoch = csv["epoch"]
    training_loss = csv["training_loss"]
    validation_loss = csv["validation_loss"]
    bleu = csv["bleu"]

    fig, ax1 = plt.subplots()
    ax_twin = ax1.twinx()

    ax1.plot(epoch, training_loss, label="training loss", color="red")
    ax1.plot(epoch, validation_loss, label="validation loss", color="blue")
    ax_twin.plot(epoch, bleu, label="bleu", color="green")
    ax1.set_ylabel("Loss")
    ax_twin.set_ylabel("SacreBLEU")
    fig.suptitle(name)
    lines, labels = ax1.get_legend_handles_labels()
    lines2, labels2 = ax_twin.get_legend_handles_labels()
    ax1.legend(lines + lines2, labels + labels2, loc=2)
    plt.show()
    fig.savefig(output_path)


```

```python
plot_loss("Loss/opus-baseline.csv","Graph/opus-bl.png","OPUS-Baseline")
```

```python
plot_loss("Loss/nllb-baseline.csv","Graph/nllb-bl.png","NLLB-Baseline")
```

```python
plot_loss("Loss/mbart-baseline.csv","Graph/mbart-bl.png","mBart-Baseline")
```
